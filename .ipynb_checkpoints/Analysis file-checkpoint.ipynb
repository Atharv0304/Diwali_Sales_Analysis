```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\nikhi\Desktop\Diwali_sales\Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
df.drop(['Status','unnamed1'] ,axis = 1 , inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace =True)
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype(int)
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns = {'Marital_Status':'Relationship'},inplace =True)
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
ax = sns.countplot(x = 'Gender' , data = df)
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
df.groupby(['Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending =False)
```

```python
sales_gen = df.groupby(['Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending =False)
sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen)
plt.show()
```

```python
ax = sns.countplot(x = 'Age Group',hue = 'Gender' , data = df)
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_age = df.groupby(['Age Group'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending =False)
sns.barplot(x = 'Age Group' , y = 'Amount', data = sales_age)
plt.show()
```

```python
sales_state = df.groupby(['State'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending =False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'State' , y = 'Orders', data = sales_state)
plt.show()
```

```python
df.columns
```

```python
sales_state = df.groupby(['State'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending =False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'State' , y = 'Amount', data = sales_state)
plt.show()
```

```python
ax = sns.countplot(x='Relationship', data=df)
sns.set(rc={'figure.figsize':(6,5)})

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Relationship','Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending =False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Relationship' , y = 'Amount',hue = 'Gender' ,data = sales_state)
plt.show()
```

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
sales_occupation = sns.countplot( x = 'Occupation' ,data = df)
for bars in sales_occupation.containers:
    sales_occupation.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Occupation'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending =False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation' , y = 'Amount' ,data = sales_state)
plt.show()
```

```python
sales_ocu = df.groupby(['Occupation'] ,as_index = False )['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x='Occupation' , y = 'Amount',data = sales_ocu)
plt.show()
```

```python
sns.set(rc = {'figure.figsize':(30,5)})
sales_occupation = sns.countplot( x = 'Product_Category' ,data = df)
for bars in sales_occupation.containers:
    sales_occupation.bar_label(bars)

plt.show()
```

```python
sales_ocu = df.groupby(['Product_Category'] ,as_index = False )['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x='Product_Category' , y = 'Amount',data = sales_ocu)
plt.show()
```

```python
sales_ocu = df.groupby(['Product_ID'] ,as_index = False )['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x='Product_ID' , y = 'Orders',data = sales_ocu)
plt.show()
```
